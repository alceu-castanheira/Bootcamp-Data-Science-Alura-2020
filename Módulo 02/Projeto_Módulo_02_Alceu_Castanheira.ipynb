```
def fillna_dados_polio_regiao(df):
  df["1995"] = pd.to_numeric(df["1995"], errors='coerce')
  df = df.fillna(value=df.mean().round(2))
  return df
```

```
def melt_df(df):
  df = df.melt(id_vars=["uf"])
  df.columns = ["uf", "ano", "cobertura_vacinal"]
  df['cobertura_vacinal'] = pd.to_numeric(df['cobertura_vacinal'], errors='coerce')
  df['ano'] = df['ano'].astype(int)
  return df
```

```
def lineplot_uf_por_regiao(df, regiao):
  
  if regiao == 'Norte':
    num_uf = 7
  elif regiao == 'Nordeste':
    num_uf = 9
  elif regiao == "Sudeste":
    num_uf = 4
  elif regiao == "Sul":
    num_uf = 3
  elif regiao == "Centro-Oeste":
    num_uf = 4
  else:
    print('Insira uma região válida: Norte, Nordeste, Sudeste, Sul ou Centro-Oeste') 

  plt.figure(figsize=(20,25))
  #plt.subplots_adjust(left = 0.125, right = 0.9, bottom = 0.1, top = 0.9, wspace = 0.2, hspace = 0.2)

  plt.subplot(311)
  sns.lineplot(data=df.iloc[0:(num_uf*7)], x="ano", y="cobertura_vacinal", hue="uf", style="uf", markers=True)
  plt.xticks(rotation=30)
  plt.xlabel('Ano', fontsize=14)
  plt.ylabel('Cobertura Vacinal (%)', fontsize=14)
  plt.title('Porcentagem de Cobertura Vacinal nas unidades da Federação da região {} entre 1994 e 2000'.format(regiao), fontsize=16)

  plt.subplot(312)
  sns.lineplot(data=df.iloc[(num_uf*7):(num_uf*7+num_uf*10)], x="ano", y="cobertura_vacinal", hue="uf", style="uf", markers=True)
  plt.xticks(rotation=30)
  plt.xlabel('Ano', fontsize=14)
  plt.ylabel('Cobertura Vacinal (%)', fontsize=14)
  plt.title('Porcentagem de Cobertura Vacinal nas unidades da Federação da região {} entre 2001 e 2010'.format(regiao), fontsize=16)

  plt.subplot(313)
  sns.lineplot(data=df.iloc[(num_uf*7+num_uf*10):], x="ano", y="cobertura_vacinal", hue="uf", style="uf", markers=True)
  plt.xticks(rotation=30)
  plt.xlabel('Ano', fontsize=14)
  plt.ylabel('Cobertura Vacinal (%)', fontsize=14)
  plt.title('Porcentagem de Cobertura Vacinal nas unidades da Federação da região {} entre 2011 e 2019'.format(regiao), fontsize=16)

  plt.show()
```

```
dados_polio_norte = melt_df(dados_polio_norte)
dados_polio_norte
```

```
dados_polio_uf_total = dados_polio_uf_total.to_frame()
dados_polio_uf_total
```

```
def separar_total_por_regiao(regiao):

  if regiao == 'Norte':
    lista_uf = ["Rondônia", "Acre", "Amazonas", "Roraima", "Pará", "Amapá", "Tocantins", "Total"]
    df = dados_polio_uf_total[0:7].append(dados_polio_uf_total.iloc[27], ignore_index=True)
  elif regiao == 'Nordeste':
    lista_uf = ["Maranhão", "Piauí", "Ceará", "Rio Grande do Norte", "Paraíba", "Pernambuco", "Alagoas", "Sergipe", "Bahia", "Total"]
    df = dados_polio_uf_total[7:16].append(dados_polio_uf_total.iloc[27], ignore_index=True)
  elif regiao == "Sudeste":
    lista_uf = ["Minas Gerais", "Espírito Santo", "Rio de Janeiro", "São Paulo", "Total"]
    df = dados_polio_uf_total[16:20].append(dados_polio_uf_total.iloc[27], ignore_index=True)
  elif regiao == "Sul":
    lista_uf = ["Paraná", "Santa Catarina", "Rio Grande do Sul", "Total"]
    df = dados_polio_uf_total[20:23].append(dados_polio_uf_total.iloc[27], ignore_index=True)  
  elif regiao == "Centro-Oeste":
    lista_uf = ["Mato Grosso do Sul", "Mato Grosso", "Goiás", "Distrito Federal", "Total"]
    df = dados_polio_uf_total[23:27].append(dados_polio_uf_total.iloc[27], ignore_index=True)  
  else:
    print('Insira uma região válida: Norte, Nordeste, Sudeste, Sul ou Centro-Oeste') 

  df['uf'] = lista_uf
  df = df.sort_values(by=['Total'], ascending=False)
  
  return df
```

```
def barplot_total_estados_por_regiao(regiao):
  dados_polio_estados_por_regiao = separar_total_por_regiao(regiao)

  plt.figure(figsize=(14,7))
  sns.barplot(x="uf",y="Total",data=dados_polio_estados_por_regiao)
```
